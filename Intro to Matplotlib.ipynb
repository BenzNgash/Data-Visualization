# Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# plot to each different axis
ax1.plot(x, np.sin(x));
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly use pyplot method
* More advanced use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
```

```python
# 00 method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)

#ax.set_xlim([45, 100])
```

```python
over_50.target.values
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50['target']);

#Customize
ax.set(title="Heart Disease and CHolestorol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
over_50.head()
```

```python
# Subplot of chol, age & thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     )
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# add legend to ax0
ax0.legend(*scatter.legend_elements("colors"), title="Target")

# add a meanline
ax0.axvline(x=over_50["age"].mean(),
           linestyle="--",
           label="Mean Age",
           color="red")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--",
           color="blue",
            label="Mean Cholesterol"
    )
# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

#Customize ax1
ax1.set(title="Heart Disease and Thalach levels",
       xlabel="Age",
       ylabel="Max heart rate")

# add legend
ax1.legend(*scatter.legend_elements(), title="Target")

# add meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--",
            color="green"
           );

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

## Matplotlib exercises

```python
import pandas as pd

# Data for temperatures and Vout
data = {
    "Temperature (C)": [30, 32, 34, 36, 38, 40],
    "Vout (V)": [0.11, 0.16, 1.09, 1.32, 1.55, 2.0]
}

# Creating the DataFrame
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

# Sample data
temperature = [30, 32, 34, 36, 38, 40]
vout = [0.11, 0.16, 1.09, 1.32, 1.55, 2.0]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(temperature, vout, marker='o', color='blue', label="Vout (V)")

# Add labels and title
plt.xlabel("Temperature (C)")
plt.ylabel("Vout (V)")
plt.title("Temperature vs Vout")

# Optional mean lines
plt.axvline(x=sum(temperature) / len(temperature), linestyle="--", color="red", label="Mean Temperature")
plt.axhline(y=sum(vout) / len(vout), linestyle="--", color="green", label="Mean Vout")

# Add legend
plt.legend()

plt.savefig("temperature_vs_vout.png", format="png", dpi=300)

# Show the plot
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
temperature = [30, 32, 34, 36, 38, 40]
vout = [0.11, 0.16, 1.09, 1.32, 1.55, 2.0]

# Fit a polynomial to the data
z = np.polyfit(temperature, vout, 3)  # 3rd-degree polynomial for smoother curve
p = np.poly1d(z)

# Generate smooth x values and corresponding y values
x_smooth = np.linspace(min(temperature), max(temperature), 100)
y_smooth = p(x_smooth)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_smooth, y_smooth, color='blue', label="Vout (V) - Smoothed Curve")
plt.scatter(temperature, vout, color='blue')  # original points

# Add labels and title
plt.xlabel("Temperature (C)")
plt.ylabel("Vout (V)")
plt.title("Temperature vs Vout")

# Optional mean lines
plt.axvline(x=np.mean(temperature), linestyle="--", color="red", label="Mean Temperature")
plt.axhline(y=np.mean(vout), linestyle="--", color="green", label="Mean Vout")

# Add legend
plt.legend()

# Save the figure
plt.savefig("temperature_vs_vout_curved.png", format="png", dpi=300)

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
# Lab data
temperature = [30, 32, 34, 36, 38, 40]
vout = [0.11, 0.16, 1.09, 1.32, 1.55, 2.0]
thermocouple_voltage = [2.71, 2.46, 2.48, 2.60, 2.65, 2.70]
rtd_voltage = [3.0, 3.2, 3.4, 3.6, 3.8, 4.0]


# Create plot
plt.figure(figsize=(10, 8))
plt.plot(temperature, vout, marker='o', color='blue', label="Vout (V)")
plt.plot(temperature, thermocouple_voltage, marker='s', color='purple', label="Thermocouple Voltage (V)")
plt.plot(temperature, rtd_voltage, marker='^', color='orange', label="RTD Voltage (V)")

# labels and title
plt.xlabel("Temperature (C)")
plt.ylabel("Voltage (V)")
plt.title("Temperature vs Vout, Thermocouple, and RTD Voltages")

# Add mean lines 
plt.axhline(y=sum(vout) / len(vout), linestyle="--", color="green", label="Mean Vout")
plt.axhline(y=sum(thermocouple_voltage) / len(thermocouple_voltage), linestyle="--", color="purple", label="Mean Thermocouple Voltage")
plt.axhline(y=sum(rtd_voltage) / len(rtd_voltage), linestyle="--", color="orange", label="Mean RTD Voltage")

# Add legend
plt.legend()

# Save the figure
plt.savefig("temperature_vs_all_voltages_1.png", format="png", dpi=300)

# Show the plot
plt.show()
```
